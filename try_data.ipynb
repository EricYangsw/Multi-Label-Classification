```python
import Cython
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from fastxml import Trainer, Inferencer
```

```python
path_x = '../data/devide_data/x/'
path_y = '../data/devide_data/y/'
```

```python
train_x = []
train_y = []

input_ = pd.read_csv(path_x + 'x_data_' + str(0) + '.csv', index_col=0).values
label  = pd.read_csv(path_y + 'y_data_' + str(0) + '.csv', index_col=0).values
for i in range(3999):
    train_x.append(csr_matrix(input_[i, :]))
    train_y.append(np.where(label[i, :]==1)[0].tolist())
                   
```

```python
trainer = Trainer(n_trees=2, n_jobs=-1)
```

```python
trainer.fit(train_x, train_y)
```

```python
save_path = './saver'
trainer.save(save_path)
```
